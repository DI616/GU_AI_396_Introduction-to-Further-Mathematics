**1.1 Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10). Найдите их сумму.**

Сумма векторов \[10, 10, 10\] и \[0, 0, -10\] равна \[10, 10, 0\]

**1.2 Напишите код на Python, реализующий расчет длины вектора, заданного его координатами.**

```python
v = [10, 10, 0]

def get_vector_len(vec):
    
    k = 0
    
    for a in vec:
        k += a**2
        
    return math.sqrt(k)

print(get_vector_len(v))
```

**2. Почему прямые не кажутся перпендикулярными?**

```python
x = np.linspace(-5, 5, 21)
y1 = 3*x+1
y2 = (-1/3)*x+1

fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.set(aspect=1)
ax1.plot(x, y1)
ax1.plot(x, y2)
ax2.plot(x, y1)
ax2.plot(x, y2)
plt.show()
```

Прямые не кажутся перпердикулярными, так как соотношения координатных прямых не 1:1

**3. Напишите код на Python, реализующий построение графиков:**
* окружности,
* эллипса,
* гиперболы.

```python
x = np.linspace(-10, 10, 1001)
y11 = np.sqrt(4 - (x)**2)
y12 = np.sqrt(4 - (x)**2) * -1

y21 = np.sqrt((1 - (x**2)/100) * 25)
y22 = np.sqrt((1 - (x**2)/100) * 25) * -1

y31 = np.sqrt((-1 + (x**2)/25) * 25)
y32 = np.sqrt((-1 + (x**2)/25) * 25) * -1

fig, ax = plt.subplots(nrows=1, ncols=3)

ax1, ax2, ax3 = ax.flatten()
ax1.set(aspect=1)
ax2.set(aspect=1)
ax3.set(aspect=1)
fig.set_size_inches(16, 5)
ax1.plot(x, y11, color='red')
ax1.plot(x, y12, color='red')
ax2.plot(x, y21, color='red')
ax2.plot(x, y22, color='red')
ax3.plot(x, y31, color='red')
ax3.plot(x, y32, color='red')
plt.show()
```

**4.2 Пусть задана плоскость: A1\*x + B1\*y + C1\*z + D1 = 0 и прямая:**

**(x-x1)/(x2-x1)=(y-y1)/(y2-y1)=(z-z1)/(z2-z1)**

**Как узнать, принадлежит прямая плоскости или нет?**

Выбрать две произвольные точки прямой и подставить их в уравнение плоскости. Если уравнение окажется верно, то прямая принадлежит плоскости.

**5.1 Нарисуйте трехмерный график двух параллельных плоскостей.**

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, .5)
Y = np.arange(-5, 5, .5)
X, Y = np.meshgrid(X, Y)
Z1 = 2*X + 5*Y
Z2 = 2*X + 5*Y + 30
ax.plot_wireframe(X, Y, Z2, color='red')
ax.plot_surface(X, Y, Z1)
show()
```

**5.2 Нарисуйте трехмерный график двух любых поверхностей второго порядка.**

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, .5)
Y = np.arange(-5, 5, .5)
X, Y = np.meshgrid(X, Y)
Z = (X**2)/5 - (Y**2)/10
ax.plot_surface(X, Y, Z)
show()
```
