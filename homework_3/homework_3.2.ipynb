**3.1 Напишите код, который будет переводить полярные координаты в декартовы.**

```python
def polar_to_decart(a, r):
    
    a = np.pi/180*a
    
    x = r*np.cos(a)
    y = r*np.sin(a)
    
    return (x, y)

polar_to_decart(45, 10)
```

**4.1 Решите систему уравнений: y = x^2 – 1, exp(x) + x*(1 – y) = 1**

```python
x = np.linspace(-5, 5, 1001)
y1 = (x**2)-1
y2 = 1 - (1 - np.exp(x))/x
plt.grid(True)
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
def equations(p):
    x, y = p
    return (y - (x**2) + 1, y - 1 + (1 - np.exp(x))/x)

x_1, y_1 = fsolve(equations, (-2, -0))
x_2, y_2 = fsolve(equations, (2, 5))
x_3, y_3 = fsolve(equations, (4, 15))

print(x_1, y_1, '\n', x_2, y_2, '\n', x_3, y_3)
```

**4.2 Решите систему уравнений и неравенств: y = x^2 – 1, exp(x) + x*(1 – y) > 1**

```python
x = np.linspace(-5, 5, 1001)
y1 = (x**2)-1
y2 = 1 - (1 - np.exp(x))/x
plt.grid(True)
plt.plot(x, y1, color='red')
plt.plot(x, y2, linestyle='--', color='orange')
plt.fill_between(x, y1, y2, where=(y1<y2), color='green')
plt.show()
```
